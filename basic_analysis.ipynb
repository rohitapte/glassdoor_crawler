```python
import json
import nltk
```

```python
data=[]
with open('Amazon.txt') as f:
    for line in f:
        data.append(json.loads(line))
len(data)
```

```python
data[0]
```

```python
corpus=[]
for item in data:
    if item['pros']!='':
        for subitem in item['pros'].split('\n'):
            corpus.append(subitem.lower())
    if item['cons']!='':
        for subitem in item['cons'].split('\n'):
            corpus.append(subitem.lower())
    if item['adviceManagement']!='':
        for subitem in item['adviceManagement'].split('\n'):
            corpus.append(subitem.lower())
len(corpus)
```

```python
corpus
```

```python
avg_words=0
max_words=0
for item in corpus:
    words=nltk.word_tokenize(item)
    if len(words)>max_words:
        max_words=len(words)
    avg_words+=len(words)
avg_words/=len(corpus)
print(avg_words)
print(max_words)
```

```python
from gensim import utils

class MyCorpus:
    """An iterator that yields sentences (lists of str)."""

    def __iter__(self):
        for line in corpus:
            # assume there's one document per line, tokens separated by whitespace
            yield utils.simple_preprocess(line)
```

```python
import gensim.models

sentences = MyCorpus()
model = gensim.models.Word2Vec(sentences=sentences)
```

```python
from collections import Counter
from nltk.corpus import stopwords

stop_words=stopwords.words('english')
stop_words.append(',')
stop_words.append('.')
stop_words.append('-')
all_words=[]
for item in corpus:
    all_words.extend([word for word in nltk.word_tokenize(item) if word not in stop_words])
word_counts=Counter(all_words)
```
