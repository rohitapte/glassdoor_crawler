```python
import json
import nltk
from collections import defaultdict
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
proCount=defaultdict(int)
conCount=defaultdict(int)
adviceManagementCount=defaultdict(int)
combinedCount=defaultdict(int)

for file in os.listdir('data'):
    with open(os.path.join('data',file),encoding='utf-8') as f:
        for line in f:
            data=json.loads(line)
            proTokens=len(nltk.word_tokenize(data['pros']))
            conTokens=len(nltk.word_tokenize(data['cons']))
            adviceManagementTokens=len(nltk.word_tokenize(data['adviceManagement']))
            proCount[proTokens]+=1
            conCount[conTokens]+=1
            adviceManagementCount[adviceManagementTokens]+=1
            combinedCount[proTokens]+=1
            combinedCount[conTokens]+=1
            combinedCount[adviceManagementTokens]+=1
```
